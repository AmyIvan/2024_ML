```python
import pandas as pd

data_path = './data/'
train_data = pd.read_csv(data_path + 'train_data.csv')
test_data = pd.read_csv(data_path + 'test_data.csv')
```

```python
# 将train_data和test_data合并为一个数据集



df = pd.concat([train_data, test_data], axis=0)

df['datetime'] = pd.to_datetime(df['dteday']) + pd.to_timedelta(df['hr'], unit='h')

# 格式化小时为 hh:00
df['date'] = df['datetime'].dt.strftime('%Y-%m-%d %H:%M')

# 删除datetime列和dteday列和hr列
df.drop(['datetime', 'dteday', 'hr', 'mnth'], axis=1, inplace=True)
# 把最后一列移动到第二列
cols = df.columns.tolist()
cols = [cols[-1]] + cols[:-1]
df = df[cols]

# 将df保存为csv文件
df = df.dropna()
df.to_csv('./formatted_data.csv', index=False)
```
